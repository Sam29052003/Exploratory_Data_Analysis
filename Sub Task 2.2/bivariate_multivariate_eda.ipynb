```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
df = pd.read_csv("cleaned_data.csv")
df.head()
```

```python
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
categorical_cols = df.select_dtypes(include=['object']).columns

numerical_cols, categorical_cols
```

```python
# BIVARIATE ANALYSIS

sns.scatterplot(x="distance_km", y="order_value", data=df)
plt.title("Distance vs Order Value")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.heatmap(df[numerical_cols].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

```python
sns.barplot(x="restaurant_type", y="order_value", data=df)
plt.title("Restaurant Type vs Order Value")
plt.xticks(rotation=30)
plt.show()
```

```python
df.groupby("city")["order_value"].mean().sort_values(ascending=False)
```

```python
sns.countplot(x="day_of_week", hue="restaurant_type", data=df)
plt.xticks(rotation=45)
plt.title("Orders by Day and Restaurant Type")
plt.show()
```

```python
# MULTIVARIATE ANALYSIS

sns.pairplot(df[numerical_cols])
plt.show()
```

```python
sns.scatterplot(
    x="distance_km",
    y="delivery_time_min",
    hue="restaurant_type",
    data=df
)
plt.title("Distance vs Delivery Time by Restaurant Type")
plt.show()
```
